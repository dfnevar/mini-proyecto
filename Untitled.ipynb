# Importo las labrerias necesarias para extraer los datos de la web

```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
import re
```

# Elijo y nombro la web con la que voy a trabajar, hago el request y la soup para extraer los datos en formato HTML

```python
url = "https://www.motogp.com/es/gp-results"
```

```python
r=requests.get(url)
```

```python
soup = BeautifulSoup(r.content,'html.parser')
soup
```

# Una vez tengo los datos ya en formato HTML extraigo todos los 'td' que segun la web son cada una de las columnas de la tabla

```python
rows=soup.find_all('td')
rows
```

# Limpio la lista de los datos obtenidos , buscando solo el texto, eliminiando los saltos de fila y eliminando los espacios en blanco 

```python
clean = [i.getText().replace("\n", "").strip() for i in rows]
clean
```

# Algunos datos dentro de la lista me quedan agrupados como un solo valor, por lo que los separo para tenerlos asi todos separados


```python
clean1=[]
for i in clean:
    elemento = i.strip().split("             ")
    if len(elemento) == 2:
        clean1.append(elemento[0])
        clean1.append(elemento[1])
    else:
        clean1.append(elemento[0])
clean1
```
